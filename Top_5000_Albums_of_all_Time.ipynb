```python
df.rename(columns = {'Unnamed: 0' : 'id_number'}, inplace = True) #Renaming column 'Unnamed: 0'
```

```python
df.set_index('id_number')      #Setting new column as index column
```

```python
df_na = df.loc[df['descs'].isna()] # Creating new DataFame where column 'descs' value equals to NaN
```

```python
df_na.head(3)
```

```python
df_na['descs'] = df_na.loc[:, ('gens')].str.split(",", 1, expand=True)  # Replacing NaN value with splitted value from column 'gens'
```

```python
df_na.isna().sum()
```

```python
df[['nvalue1', 'nvalue2']] = df['gens'].str.split(",", 1, expand=True)  # Creating new columns by splitting column 'gens' by first coma
```

```python
df.head(3)
```

```python
df['descs'] = df['descs'].fillna(df['nvalue1'])   # Replacing NaN value with value from column 'nvalue1'
```

```python
df.isna().sum()
```

```python
df.drop('nvalue2', axis=1) # Deleting column 'nvalue2'
```

```python
df['Year'] = pd.to_datetime(df['rel_date']).dt.year     #Creating new column 'Year'
```

```python
df.head(3)
```

```python
df.groupby('Year').agg(Release_Total=('Year', 'count')).reset_index()
```

```python
df.groupby('Year').agg(Release_Total=('Year', 'count'),min_AVG_rating=('avg_rat','min'),max_AVG_rating=('avg_rat', 'max'),min_NUM_rat=('num_rat','min'),max_NUM_rat=('num_rat','max')).reset_index()
```

```python
plt.figure(figsize=(22,5))
chart = sns.countplot(x='Year', data = df, palette='GnBu')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',   #escaping overlapping
    fontweight='light')

None   # don't show the label objects
```

###  The most number of Albums of All Times were released in the years 1971, 1972 and 1973.

```python
plt.figure(figsize=(22,5))
chart = sns.countplot(x='nvalue1', data = df[df['Year'].isin([1971])], palette='OrRd')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light')
chart.set(xlabel='Genre')

None 
```

###  Y1971, top 3 the most popular genres were Progressive Rock, Hard Rock, and Blues Rock.

```python
by_genre = (df
            .groupby('nvalue1')
            .filter(lambda x : len(x) > 20) # filtering more than 20 albums
            .groupby(['nvalue1', 'Year'])
            .size()
            .unstack()
           )
```

```python
plt.figure(figsize=(20,20))
hmap = sns.heatmap(
    by_genre, 
    square=True, # make cells square
    cbar_kws={'fraction' : 0.01}, # shrink colour bar
    cmap='PuRd', # use purple/red colour map
    linewidth=1 # space between cells
)
```

### The first most popular genre was Hard Bop, from Y1957 till Y1965.
### Progressive Rock was popular from Y1970 till Y1977.
### In Y1977 Punk Rock and Art Pop were popular as well.
### In last 10 years non of genres is most popular.

```python
plt.figure(figsize=(22,5))
chart = sns.scatterplot(x = 'nvalue1', y = 'avg_rat', data = df[df['Year'].isin([1971])])
plt.xticks(
    rotation=90, 
    horizontalalignment='right',
    fontweight='light')
chart.set(xlabel='Genre')
chart.set(ylabel='Average Rating')
plt.show()
```
